# Machine Learning: Introduction, Supervised Learning (Classification & Regression) & Linear Regression Fundamentals

## What is Machine Learning?

* **Definition (Arthur Samuel):** Field of study that gives computers the ability to learn without being explicitly programmed.
* **Key Idea:** The more opportunities (data/experience) a learning algorithm has, the better it performs.
* **Core Principle:** Programs learn from data to get better at tasks over time.
* **Practical Advice Emphasis:** This course focuses on both the tools (algorithms) and the skills to **apply** them effectively to build valuable ML systems.

## Supervised Learning: Input to Output Mappings (X to Y)

* **Definition:** Algorithms that learn **input (X) to output (Y) mappings** by being given examples with **"right answers"** (correct labels/outputs).
* **Process:** Train with (X, Y) pairs, then predict Y for new, unseen X.
* **Economic Value:** ~99% of current ML economic value is from supervised learning.

### Examples of Supervised Learning

* **Spam Filter:**
    * Input (X): Email
    * Output (Y): Spam / Not Spam
* **Speech Recognition:**
    * Input (X): Audio clip
    * Output (Y): Text transcript
* **Machine Translation:**
    * Input (X): English text
    * Output (Y): Corresponding Spanish/Hindi/etc. text
* **Online Advertising:**
    * Input (X): Ad info, User info
    * Output (Y): Will user click on ad (Yes/No)?
* **Self-Driving Cars:**
    * Input (X): Image, Sensor data (radar)
    * Output (Y): Position of other cars
* **Manufacturing (Visual Inspection):**
    * Input (X): Picture of manufactured product (e.g., cell phone)
    * Output (Y): Defect (Scratch/Dent/None)?

---

### 1. Regression

* **Definition:** Predicts a **numerical output** from an **infinite number of possible values**.
* **Example:** Housing Price Prediction
    * Input (X): Size of house (e.g., in sq ft)
    * Output (Y): Price of house (e.g., in thousands of dollars)
    * Model can fit a straight line or a curve to the data.

---

### 2. Classification

* **Definition:** Predicts a **category** or a **small, finite set of possible discrete outputs**.
* **Examples:**
    * **Breast Cancer Detection:**
        * Input (X): Tumor size, Patient age, Tumor clump thickness, Cell uniformity.
        * Output (Y): Malignant (1) / Benign (0)
        * Can also have more than two output categories (e.g., Cancer Type 1, Type 2, Benign).
    * **Image Recognition:** Predict if a picture is a cat or a dog.
    * **Disease Diagnosis:** Predict if a patient has a specific disease (e.g., Diabetes / Not Diabetes).

* **Key Difference from Regression:** Classification outputs are discrete categories (e.g., 0, 1, or "cat", "dog"), not continuous numbers (e.g., 0.5, 1.7).

---


## Unsupervised Learning

* **Definition:** Data comes **only with inputs (X), no output labels (Y)**. The algorithm's job is to find some **structure, pattern, or something interesting** in the data on its own.
* **Purpose:** Discover hidden groupings or patterns.

---

### Types of Unsupervised Learning

1.  **Clustering (Most Common):**
    * **Goal:** Groups similar data points into different **clusters** or categories.
    * **Examples:**
        * **Google News:** Groups hundreds of thousands of news articles by related stories (e.g., based on similar words like "panda," "twin," "zoo").
        * **Genetic / DNA Data:** Groups individuals into different categories based on gene expression patterns.
        * **Customer Market Segmentation:** Groups customers into distinct market segments based on their behavior or motivations (e.g., seeking knowledge, career development, staying updated).

2.  **Anomaly Detection:**
    * **Goal:** Detects **unusual events** or outliers.
    * **Application:** Fraud detection in financial systems (unusual transactions).

3.  **Dimensionality Reduction:**
    * **Goal:** Compresses a **large dataset into a much smaller dataset** while losing as little information as possible.
    * **Application:** Useful for simplifying complex data for analysis or visualization.

## Supervised Learning Process: Linear Regression Model


---

### The Linear Regression Model

* **Problem:** Predict housing prices based on house size.
* **Dataset:** House sizes (sq ft) and prices (thousands of dollars).
* **Model Goal:** Fit a **straight line** to the data to make predictions.
* **Notation:**
    * **Training Set:** Dataset used to train the model.
    * **Input Variable (Feature):** `x` (e.g., house size).
    * **Output Variable (Target):** `y` (e.g., house price).
    * **Number of Training Examples:** `m` (e.g., 47 houses).
    * **Single Training Example:** $(x, y)$
    * **i-th Training Example:** $(x^{(i)}, y^{(i)})$
        * **Note:** Superscript `(i)` is an index, not an exponentiation.

* **Model Function (`f`):**
    * Takes input $x$ and outputs a **prediction $\hat{y}$**.
    * $\hat{y}$ (y-hat) is the estimated/predicted value of $y$.
    * $f$ is also called the **model** or historically, **hypothesis**.
    * For linear regression with one variable: $f_{w,b}(x) = w \cdot x + b$
        * $w$ and $b$ are **parameters** (also called coefficients or weights).
        * $w$: Controls the **slope** of the line.
        * $b$: Controls the **y-intercept** (where the line crosses the vertical axis).
        * Often simplified to $f(x) = w \cdot x + b$.

* **Linear Regression with One Variable (Univariate Linear Regression):**
    * Refers to a model with a single input variable/feature $x$.

---

## Cost Function (J)

* **Purpose:** Measures **how well the model's predictions ($\hat{y}$) align with the actual target values ($y$)** in the training data.
* **Goal:** Choose $w$ and $b$ to **minimize** the cost function $J(w, b)$.

---

### Squared Error Cost Function

* **Error for a single example $i$:** $(f(x^{(i)}) - y^{(i)})$
* **Squared Error for a single example $i$:** $(f(x^{(i)}) - y^{(i)})^2$
* **Formula ($J(w, b)$):**
    $$ J(w, b) = \frac{1}{2m} \sum_{i=1}^{m} (f_{w,b}(x^{(i)}) - y^{(i)})^2 $$
    * $m$: Number of training examples.
    * $1/2m$: Divides by $2m$ to compute the **average squared error** and simplify later derivative calculations.
* **Properties:**
    * For linear regression, the squared error cost function is always **convex (bowl-shaped)**.
    * This means it has a **single global minimum**, no multiple local minima.

---


```python
# Example of a simple cost function (for intuition, not direct use in full linear regression)
import numpy as np
import matplotlib.pyplot as plt

def simple_cost_function(w, x_data, y_data):
    """
    Calculates the cost for a simplified model f(x) = w * x
    """
    m = len(x_data)
    cost = 0
    for i in range(m):
        prediction = w * x_data[i]
        error = prediction - y_data[i]
        cost += error**2
    return cost / (2 * m)

# Example data points (1,1), (2,2), (3,3)
x_train = np.array([1, 2, 3])
y_train = np.array([1, 2, 3])

# Calculate cost for different 'w' values
w_values = np.linspace(-1, 3, 50)
costs = [simple_cost_function(w, x_train, y_train) for w in w_values]

# Plot the cost function J(w)
plt.figure(figsize=(8, 6))
plt.plot(w_values, costs, '-b')
plt.xlabel('Parameter w')
plt.ylabel('Cost J(w)')
plt.title('Cost Function J(w) for simplified Linear Regression (f(x) = w*x)')
plt.grid(True)
plt.show()

# You can add more code cells here to demonstrate a simple gradient descent step
# or visualize different lines with different w,b values if you choose to.
# For full linear regression, you'd implement the gradient descent algorithm
# with the derivatives for both w and b.
```

### Visualizing the Cost Function (J)

* **1 Parameter (Simplified: $f(x) = w \cdot x$, so $b=0$):**
    * $J(w)$ is a **U-shaped curve** (like a soup bowl in 2D).
    * The minimum of $J(w)$ corresponds to the $w$ value where the line $f(x)$ best fits the data.

* **2 Parameters ($w$, $b$):**
    * $J(w, b)$ is a **3D bowl-shaped surface**.
    * **Surface Plot:** Shows the height of $J$ for different $w$ and $b$ values.
    * **Contour Plot:** Shows concentric ellipses (ovals). Each ellipse represents points $(w, b)$ that have the **same cost value $J$**.
        * The **center of the smallest ellipse** is the global minimum of the cost function.
* **Relationship:** A better-fitting line $f(x)$ (visualized on the data plot) corresponds to a point $(w, b)$ on the $J$ graph that is **closer to the minimum**.

## Gradient Descent Algorithm

* **Purpose:** An efficient algorithm to **automatically find the values of parameters ($w$, $b$) that minimize the cost function $J$**.
* **Core Idea:** Start with initial parameter guesses, then repeatedly adjust them by taking steps in the direction of **steepest descent** (downhill) until a minimum is reached.
* **Applicability:** Used to train many ML models, including advanced neural networks (deep learning).

---

### Gradient Descent Update Rule

* **Simultaneous Update:**
    $$ w := w - \alpha \frac{\partial}{\partial w} J(w, b) $$
    $$ b := b - \alpha \frac{\partial}{\partial b} J(w, b) $$
    * **`:=` (Assignment Operator):** Means "set $w$ (or $b$) to the new calculated value."
    * **$\alpha$ (Alpha): Learning Rate:**
        * A small positive number (e.g., 0.01).
        * Controls the **size of the steps** taken downhill.
        * **Too small:** Slow convergence (many tiny steps).
        * **Too large:** May overshoot the minimum, fail to converge, or even diverge (cost increases).
    * **$\frac{\partial}{\partial w} J(w, b)$ (Derivative Term):**
        * Indicates the **direction of steepest descent** (slope of the tangent line).
        * Also influences the step size (larger derivative = steeper slope = larger step).
        * As it approaches a minimum, the derivative approaches zero, causing the update steps to automatically become smaller.

* **Simultaneous Update Logic (Crucial for Correctness):**
    1.  Calculate both derivative terms using the *current* $w$ and $b$.
    2.  Store the *new* $w$ and $b$ values in temporary variables (e.g., `temp_w`, `temp_b`).
    3.  Update $w$ and $b$ *simultaneously* using the values from the temporary variables.

---

### Derivatives for Linear Regression (Squared Error Cost)

* **Derivative with respect to $w$:**
    $$ \frac{\partial}{\partial w} J(w, b) = \frac{1}{m} \sum_{i=1}^{m} (f_{w,b}(x^{(i)}) - y^{(i)})x^{(i)} $$
* **Derivative with respect to $b$:**
    $$ \frac{\partial}{\partial b} J(w, b) = \frac{1}{m} \sum_{i=1}^{m} (f_{w,b}(x^{(i)}) - y^{(i)}) $$
    * **Note:** These derivations use calculus but can be used directly without understanding the calculus if preferred.

---

### Batch Gradient Descent

* **Definition:** On every step of gradient descent, the algorithm looks at **all ($m$) training examples** to compute the derivatives and update parameters.

---

### Gradient Descent in Action (Visualization)

* Initial $w$ and $b$ (e.g., $w = -0.1$, $b = 900$) correspond to a poor-fitting line and a high cost.
* With each step of gradient descent, the parameters $w$ and $b$ adjust, causing:
    * The line $f(x)$ to get progressively **closer to fitting the data**.
    * The cost $J(w, b)$ to **decrease** at each update, moving closer to the global minimum on the contour plot.
* The process continues until $w$ and $b$ converge at the **global minimum**, where the line $f(x)$ provides the best fit for the training data.
